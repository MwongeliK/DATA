```python
import numpy as np
import pandas as pd
```

```python
#Loading the dataset
df_csv = pd.read_csv("productdata.csv")
```

```python
df_csv.head()
```

```python
df_csv.info()
```

```python
#Check missing values
print(df_csv.isnull().sum())

#Drpping rows that have missing critical columns
df_csv = df_csv.dropna(subset=["PRODUCTID"])
```

```python
print(df_csv.duplicated().sum())
```

```python
#Removing duplicates
df_csv= df_csv.drop_duplicates().reset_index(drop=True)
print("duplicate rows:",df_csv.duplicated().sum())

```

```python
df_csv = df_csv.reset_index(drop=True)
```

```python
#standardizing ....renaming columns for consistency
df_csv.columns = df_csv.columns.str.lower().str.replace(" ","_")
df_csv.columns
```

```python
import re
```

```python
def generate_short_title(title):
    #removing unnecessary words
    if pd.isna(title):
        return ""

    title = re.sub(r"\b(set of|includes|features|with| for)\b","",title ,flags=re.IGNORECASE)
    title = re.sub(r"[^\w\s-]","",title) # removing special characters
    words = title.split()

    #keepimg the most relevant parts and limiting to 30-50 characters
    return " ".join(words[:5]) if len(title)>50 else title

    #Applying to dataset
    df_csv["short-title"] = df_csv["product_title"].apply(generate_short_title)
```

```python
df_csv.to_csv("cleaned_product_data.csv", index=False)
```
